# E-commerce gender prediction:
### Objective
With the evolution of the information and communication technologies and the rapid growth of the Internet for the exchange and distribution of information, Electronic Commerce (e-commerce) has gained massive momentum globally, and attracted more and more worldwide users overcoming the time constraints and distance barriers.

It is important to gain in-depth insights into e-commerce via data-driven analytics and identify the factors affecting product sales, the impact of characteristics of customers on their purchase habits.

It is quite useful to understand the demand, habits, concern, perception, and interest of customers from the clue of genders for e-commerce companies.

However, the genders of users are in general unavailable in e-commerce platforms. To address this gap the aim here is to predict the gender of e-commerceâ€™s participants from their product viewing records.

#### Feature extraction:
We need to extract features from train dataset, here each product consists of category/sub_category/sub_sub_category and consecutive product is separated by semi colon. We also need create a new column of time, which is the difference of time spent by user in viewing the product.

```python
def parse(data):
    pro=data['ProductList']
    time=data['total_time']
    p_enc,t,gen=[],[],[]
    if 'gender' in data:
        for i in range(0,len(pro)):
            temp=pro.iloc[i].split(';')
            for j in range(0,len(temp)):
                p_enc.append(temp[j].split('/')) 
                t.append(time[i]/len(temp))
                gen.append(data['gender'][i])
        p_id=np.array(p_enc)
        t_t=np.array(t)
        p_id=pd.DataFrame(p_id[:,0:4])
        p_id['time']=t_t
        p_id['gender']=gen
    else:
        for i in range(0,len(pro)):
            temp=pro.iloc[i].split(';')
            for j in range(0,len(temp)):
                p_enc.append(temp[j].split('/')) 
                t.append(time[i]/len(temp))
        p_id=np.array(p_enc)
        t_t=np.array(t)
        p_id=pd.DataFrame(p_id[:,0:4])
        p_id['time']=t_t        
    return p_id

def features(data):
    # Total brousing time 
    t_s=data['startTime']
    t_e=data['endTime']
    a=[]
    for i in range(0,len(t_s)):
        t1=datetime.strptime(t_e[i],'%d/%m/%y %H:%M').strftime('%Y-%m-%d %H:%M:%S.%f')
        t2=datetime.strptime(t_s[i],'%d/%m/%y %H:%M').strftime('%Y-%m-%d %H:%M:%S.%f')    
        k=datetime.strptime(t1,'%Y-%m-%d %H:%M:%S.%f')-datetime.strptime(t2,'%Y-%m-%d %H:%M:%S.%f')
        a.append(k.seconds)        
    data['total_time']=a   
    p_list=data['ProductList']  
    com_pro_feat=parse(data)
    data=data.drop(['startTime', 'endTime', 'ProductList'],axis=1)
    com_pro_feat=com_pro_feat.reset_index(drop=True)
    return com_pro_feat
```

**'features'** function will be used to convert the raw data into parsed dataset.So, this function basically parses the product list with each product in a row, gender and total time spent by the customer.
**Example:** If a male customer has viewed 5 products in a session and the total time spent by customer is 60 seconds, then the function will create 5 individual rows of male gender and time as 20 seconds in each row.

```python
train_feat=features(train) # Feature extracted from raw train data
test_feat=features(test) # Feature extracted from raw test data
train_feat.head()
```

Once data is converted into each product as row, we need to use **Label_encoder** to encode the product list. For encoding we need to concatenate both the product from train and test dataset.

```python
all_data=pd.concat([train_feat[test_feat.columns],test_feat]) # concatenate train and test data for encoding
le1 = LabelEncoder() # Encoder for category
le2 = LabelEncoder() # Encoder for sub_category
le3 = LabelEncoder() # Encoder for sub_sub_category
le4 = LabelEncoder() # Encoder for product
le5 = LabelEncoder() # Encoder for gender

le1.fit(all_data[all_data.columns[0]])
le2.fit(all_data[all_data.columns[1]])
le3.fit(all_data[all_data.columns[2]])
le4.fit(all_data[all_data.columns[3]])
le5.fit(train_feat[train_feat.columns[5]])

com_feat_train=pd.DataFrame()
com_feat_train['F1']=le1.transform(train_feat[0])
com_feat_train['F2']=le2.transform(train_feat[1])
com_feat_train['F3']=le3.transform(train_feat[2])
com_feat_train['F4']=le4.transform(train_feat[3])
com_feat_train['Time']=train_feat['time']
com_feat_train['Gender']=le5.transform(train_feat['gender'])

# Test transformed features
com_feat_test=pd.DataFrame()
com_feat_test['F1']=le1.transform(test_feat[0])
com_feat_test['F2']=le2.transform(test_feat[1])
com_feat_test['F3']=le3.transform(test_feat[2])
com_feat_test['F4']=le4.transform(test_feat[3])

com_feat_train.head()
```

Now all features are encoded, next step is to develop the classification model

#### **First method**:
**Keras** based **Artificial Neural Network** model is built on train dataset. Train dataset is split as 80% as training data and 20% as test data, training data is further split as 80% for model training and remaining 20 for validation.

```python
#%% Keras sequential cassifier. Seperate training and validation data, and scale the data
x=com_feat_train[['F1','F2','F3','F4']]
y=com_feat_train['Gender']
x_train,x_test,y_train,y_test=train_test_split(x.values,y,test_size=0.8,random_state=10)
scaler=StandardScaler().fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)

# one hot encode outputs
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Further split training dataset into model training and validation
x_train_S1,x_test_S1,y_train_S1,y_test_S1=train_test_split(x_train,y_train,test_size=0.1,random_state=32)

#% Keras deep network model
ncol = x_train.shape[1]
model=Sequential()
model.add(Dense(500, activation='linear',input_dim=ncol))
model.add(Dense(250, activation='relu'))
model.add(Dense(125, activation='relu'))
model.add(Dense(2, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
AE = model.fit(x_train_S1, y_train_S1, epochs = 50,
                    validation_data = (x_test_S1, y_test_S1)) 
training_loss = AE.history['loss']
test_loss = AE.history['val_loss']
plt.figure()
plt.plot(training_loss, 'r--')
plt.plot(test_loss, 'b-')
plt.legend(['Training Loss', 'Validation Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training & Validation loss Vs Epoch')

pred_train= model.predict(x_train)
scores = model.evaluate(x_train, y_train, verbose=0)
print('Accuracy on training data: {}% \n Error on training data: {}'.format(scores[1], 1 - scores[1]))   
 
pred_test= model.predict(x_test)
scores2 = model.evaluate(x_test, y_test, verbose=0)
print('Accuracy on test data: {}% \n Error on test data: {}'.format(scores2[1], 1 - scores2[1]))   
```

Model training is carried with 'time' column and without 'time' column and seen that there is hardly any change in accuracy of train and test data. So finally, model is trained without 'time' column.
